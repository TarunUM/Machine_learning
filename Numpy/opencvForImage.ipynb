```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
Image = cv2.imread("doctor-strange.jpg")
```

```python
plt.imshow(Image)
#open-CV reads Image/arrays iin reverese order(B, G, R)
```

```python
newImage = cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newImage)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(Image)
```

```python
faces
```

```python
x,y,w,h = faces[1]
```

```python
rgbFace = newImage[y:y+h, x:x+w]
```

```python
plt.imshow(rgbFace)
```

```python
out = cv2.rectangle(newImage, (x,y), (x+w,y+h), (0,0,255), 10)
plt.imshow(out)
```
