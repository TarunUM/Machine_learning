```python
import seaborn as sns
```

```python
x = np.random.randn(1000)
```

```python
#Histrogram Distribution
plt.hist(x, bins=30)
```

```python
#KDE(Kernal Density Estimate) Distribution 
sns.distplot(x, bins=30, kde=False, rug=True)
```

```python
#KDE(Kernal Density Estimate) Distribution with displot since distplot is deprecated
sns.displot(x, bins=30, kde=True, rug=True)
```
