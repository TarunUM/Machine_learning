```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/quotes/tag/{}?page={}"
```

```python
def get_Quotes(url):

    res=requests.get(url)

    soup = BeautifulSoup(res.text)

    quote_divs = soup.find_all("div", attrs={"class" : "quote"})

    quotes = []
    for quote_div in quote_divs:

        quote_text_div = quote_div.find_next("div", attrs={"class" : "quoteText"})

        striped = quote_text_div.text.strip()

        striped_li = striped.split("\n")

        quote = striped_li[0][1:-1]
        author = striped_li[-1].strip()

        left_div = quote_div.find_next("div", attrs={"class":"left"})
        tags = [ tag.text for tag in left_div.find_all("a") ]

        quote_item = {
            "quote" : quote,
            "authors" : author,
            "tags" : tags
        }
        quotes.append(quote_item)
    return quotes
```

```python
emotions = ["love","life","inspirational"]
total = []
for emotion in emotions:
    for i in range(1,7):
        total.extend(get_Quotes(url.format(emotion,i)))
        print(emotion," processed ",i)
```
