```python
#linspace = linear Space to get/print data
x=np.linspace(0,np.pi,100)
```

```python
y = np.sin(x) #sin takes arrays as input so all the value inside it will be stored in y array
```

```python
import matplotlib.pyplot as plt
```

```python
# 0 to 180degree of sin graph
#sin(0) = 0, sin(90) = 1 and sin(180) = 0
plt.plot(x,y)
```

```python
#to print both sin and cos graph in -90 to 90  i.e. pi/2 = 90
x = np.linspace(-np.pi/2, np.pi/2, 100)
# y = for sin value 
y = np.sin(x)
# z = for cos value 
z = np.cos(x)
```

```python
plt.plot(x,y), plt.plot(x,z)
#cos(-90) = 0 and sin(-90) = -1 
#cos(0) = 0 and sin(0) = 0
#cos(90) = 0 and sin(90) = 1
#cos is indecated in orange color line
```

```python
#for creating circle the formule is x=r sin(thetha), y=r cos(thetha)
r=3 #radius
thetha = np.linspace(0,2*np.pi,100) #2*np.pi = 360 
x=r*np.sin(thetha)
y=r*np.cos(thetha)
z=1+x 
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y), plt.scatter(z,y)
```
